```python
#author(s): Lucas Almeida Salvador e Mayara Brígida Medeiros da Silva 
'''Compara notas do Enem com infraestrutura básica 
Base de dados usada: API dados aberto educação 
Link:http://educacao.dadosabertosbr.com/api'''

import requests 
import json 
import pandas as ps 
import matplotlib.pyplot as plt
import numpy as np 

def barplotOther(agua, energia, lixo, esgoto, regiao,cor):
    values = (agua, energia, lixo, esgoto)
    n_groups = 4
    
    fig, ax = plt.subplots()
    
    index = np.arange(n_groups)
    bar_width = 0.35
    
    opacity = 0.4
    error_config = {'error': '0.3'}
    
    graph = ax.bar(index, values, bar_width, alpha = opacity, color=cor, error_kw = error_config)
    
    

    
    ax.set_xlabel = ('Infraestrutura basica')
    ax.set_ylabel = ('Porcentagem na região')
    ax.set_xticks(index + bar_width / 10)
    ax.set_xticklabels(('Agua', 'Energia', 'Coleta de lixo', 'Esgoto', 'Computadores','Internet'))
    ax.set_title(f'Gráfico porcentagem de escola com infraestrutura básica na Região {regiao}')
    
    fig.tight_layout()
    plt.show()
    
    
    
#Região Nordeste
url = "http://educacao.dadosabertosbr.com/api/estatisticas?tipoLocal=REG&nomeLocal=Nordeste&codMunicipio=0&indice=0"
data = requests.get(url).content
data = json.loads(data)
print('Nordeste')
print(f"Media enem:{data['enemMediaGeral']}")
barplotOther(data['aguaPublica']*100,data['energiaPublica']*100,data['lixoColetaPeriodica']*100,data['esgotoPublico']*100, 'Nordeste','g')


#Região Norte
url = "http://educacao.dadosabertosbr.com/api/estatisticas?tipoLocal=REG&nomeLocal=Norte&codMunicipio=0&indice=0"
data = requests.get(url).content
data = json.loads(data)

print('Norte')
print(f"Media enem:{data['enemMediaGeral']}")
barplotOther(data['aguaPublica']*100,data['energiaPublica']*100,data['lixoColetaPeriodica']*100,data['esgotoPublico']*100, 'Norte',"brown")


#Região Sul
url = "http://educacao.dadosabertosbr.com/api/estatisticas?tipoLocal=REG&nomeLocal=Sul&codMunicipio=0&indice=0"
data = requests.get(url).content
data = json.loads(data)

print('Sul')
print(f"Media enem:{data['enemMediaGeral']}")
barplotOther(data['aguaPublica']*100,data['energiaPublica']*100,data['lixoColetaPeriodica']*100,data['esgotoPublico']*100, 'Sul',"r")

#Região Sudeste
url = "http://educacao.dadosabertosbr.com/api/estatisticas?tipoLocal=REG&nomeLocal=Sudeste&codMunicipio=0&indice=0"
data = requests.get(url).content
data = json.loads(data)

print('Sudeste')
print(f"Media enem:{data['enemMediaGeral']}")
barplotOther(data['aguaPublica']*100,data['energiaPublica']*100,data['lixoColetaPeriodica']*100,data['esgotoPublico']*100, 'Sudeste',"y")

#Centro-Oeste
url = "http://educacao.dadosabertosbr.com/api/estatisticas?tipoLocal=REG&nomeLocal=Centro-Oeste&codMunicipio=0&indice=0"
data = requests.get(url).content
data = json.loads(data)

print('Centro-Oeste')
print(f"Media enem:{data['enemMediaGeral']}")
barplotOther(data['aguaPublica']*100,data['energiaPublica']*100,data['lixoColetaPeriodica']*100,data['esgotoPublico']*100, 'Centro-Oeste',"b")
```
